# Import Library

```python
import pandas as pd
```

# Data Loading

## BPJS Ketenagakerjaan

```python
df_ketenagakerjaan = pd.read_csv('ketenagakerjaan/ketenagakerjaan_clean.csv')
df_ketenagakerjaan.head()
```

## BPJS Kesehatan

### Website

```python
df_kesehatan = pd.read_csv('kesehatan/kesehatan_clean.csv')
df_kesehatan.head()
```

### Mobile

```python
df_jkn_mobile = pd.read_csv('kesehatan-jkn-mobile/jkn_mobile_clean.csv')
df_jkn_mobile.head()
```

# Combine All Dataset

```python
df = pd.concat([df_ketenagakerjaan, df_kesehatan, df_jkn_mobile])
df = df.reset_index(drop=True)
# Display all value in column
# pd.set_option('display.max_rows', None)  # Menampilkan semua baris
# pd.set_option('display.max_colwidth', None) # Menampilkan secara lengkap lebar
display(df)
```

# Save dataset final to CSV

```python
df.to_csv("final_dataset_clean.csv", index=False)
```
