#### Data JKN Mobile diambil secara manual dari FAQ aplikasi JKN Mobile

```python
import pandas as pd

df = pd.read_csv('jkn_mobile_raw.csv') 
df
```

```python
# Fungsi mapping
def map_category_subcategory(question):
    '''
    Fungsi ini memetakan pertanyaan ke kategori dan subkategori berdasarkan kata kunci
    yang sesuai dengan konteks layanan Mobile JKN.
    '''
    q = question.lower()

    # KATEGORI: KEPEESERTAAN
    if "daftar" in q or "pendaftaran" in q or "mendaftar" in q:
        if "bekerja" in q or "ppu" in q:
            return "KEPESERTAAN", "Pendaftaran PPU"
        elif "bukan pekerja" in q or "pbpu" in q:
            return "KEPESERTAAN", "Pendaftaran PBPU / BP"
        elif "bayi" in q:
            return "KEPESERTAAN", "Pendaftaran Bayi"
        elif "pensiun" in q or "veteran" in q or "ppnpn" in q or "tni" in q or "polri" in q:
            return "KEPESERTAAN", "Pendaftaran Peserta Khusus"
        elif "perusahaan" in q or "usaha" in q:
            return "KEPESERTAAN", "Pendaftaran Badan Usaha"
        return "KEPESERTAAN", "Pendaftaran Umum"

    # KATEGORI: PERUBAHAN DATA
    elif "perubahan" in q or "ubah" in q or "mengganti" in q:
        if "faskes" in q or "fasilitas kesehatan" in q:
            return "KEPESERTAAN", "Perubahan Faskes"
        elif "alamat" in q or "nomor" in q or "email" in q:
            return "KEPESERTAAN", "Perubahan Data Kontak"
        elif "kelas" in q:
            return "KEPESERTAAN", "Perubahan Kelas Rawat"
        elif "segmen" in q:
            return "KEPESERTAAN", "Perubahan Jenis Kepesertaan"
        return "KEPESERTAAN", "Perubahan Data Umum"

    # KATEGORI: IURAN & PEMBAYARAN
    elif "iuran" in q or "membayar" in q or "denda" in q or "autodebet" in q or "virtual account" in q:
        return "PEMBAYARAN", "Iuran dan Tata Cara Pembayaran"
    
    elif "tagihan" in q or "tidak sesuai" in q or "keberatan" in q:
        return "PEMBAYARAN", "Masalah Tagihan"

    # KATEGORI: KARTU & AKSES
    elif "kartu" in q or "hilang" in q or "rusak" in q or "aktifkan" in q:
        return "KARTU", "Pengelolaan Kartu Peserta"

    # KATEGORI: FASILITAS KESEHATAN
    elif "faskes" in q or "fasilitas kesehatan" in q or "tingkat pertama" in q or "rujukan" in q:
        return "PELAYANAN KESEHATAN", "Fasilitas Kesehatan"

    # KATEGORI: PELAYANAN
    elif "rawat" in q or "inap" in q or "jalan" in q or "ambulan" in q or "darurat" in q:
        return "PELAYANAN KESEHATAN", "Pelayanan Kesehatan"

    elif "klaim" in q or "refund" in q or "pengajuan" in q:
        return "PELAYANAN KESEHATAN", "Klaim & Pengembalian Dana"

    # KATEGORI: HAK & KEWAJIBAN
    elif "hak" in q or "kewajiban" in q:
        return "UMUM", "Hak dan Kewajiban Peserta"

    # KATEGORI: MOBILE JKN & DIGITAL
    elif "mobile jkn" in q or "aplikasi" in q or "e-dabu" in q or "registrasi pengguna" in q or "download" in q or "lupa password" in q:
        return "LAYANAN DIGITAL", "Aplikasi Mobile JKN"

    # KATEGORI: PENONAKTIFAN / PHK
    elif "phk" in q or "penjaminan 6 bulan" in q or "meninggal" in q or "penonaktifan" in q:
        return "STATUS KEPESERTAAN", "Penonaktifan / PHK / Meninggal"

    # KATEGORI: UMUM
    elif "apa itu" in q or "apa yang dimaksud" in q or "pengertian" in q or "bpjs kesehatan" in q:
        return "UMUM", "Definisi & Umum"

    elif "fraud" in q:
        return "UMUM", "Kecurangan Program"

    elif "layanan" in q or "call center" in q or "kontak" in q:
        return "UMUM", "Kanal Layanan"

    # Default fallback
    return "UMUM", "Lainnya"


# Buat dan isi kolom category dan subcategory berdasarkan hasil mapping dari kolom question
df[['category', 'subcategory']] = df['question'].apply(
    lambda q: pd.Series(map_category_subcategory(q))
)

# Simpan hasil ke CSV
df.to_csv("jkn_mobile_clean.csv", index=False)
```
